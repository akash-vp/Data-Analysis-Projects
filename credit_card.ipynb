```python
import os
path=("D:/Projects/Python Project ML/Machine Learning Project1")
os.chdir(path)
os.listdir(path)

import numpy as np
import pandas as pd


credit=pd.read_csv("creditcard.csv")
credit.head()


#list the columns names
print(credit.columns)
credit.describe()

#Data cleaning
credit.isnull().sum()
np.isnan(credit).sum()

#data does not contain any NULL  or missing values
#data does not contain any categorical value/column

no_of_fraud=len(credit[credit["Class"]==1])
no_of_fraud

no_of_normal=len(credit[credit["Class"]==0])
no_of_normal


x=credit.iloc[:,1:29]
print(x.head())

y=credit['Class']
print(y)


#splitting train and test data
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report


x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=10)
print(x_test.shape)

#building model
C=1.0 #SVM regularization parameter

svm_model=svm.SVC(kernel='linear',C=C,decision_function_shape='ovr').fit(x_train,y_train)
pred=svm_model.predict(x_test)
print(confusion_matrix(y_test,pred))
print(accuracy_score(y_test,pred))

report=classification_report(y_test,pred)
print(report)

from sklearn.metrics import average_precision_score, auc, roc_curve, precision_recall_curve
average_precision = average_precision_score(y_test,pred)

print('Average precision-recall score RF: {}'.format(average_precision))

#ROC curve
from sklearn.metrics import precision_recall_curve
import matplotlib.pyplot as plt

precision, recall, _ = precision_recall_curve(y_test,pred)

plt.step(recall, precision, color='b', alpha=0.2,
         where='post')
plt.fill_between(recall, precision, step='post', alpha=0.2,
                 color='b')

plt.xlabel('Recall')
plt.ylabel('Precision')
plt.ylim([0.0, 1.05])
plt.xlim([0.0, 1.0])
plt.title('2-class Precision-Recall curve: AP={0:0.2f}'.format(
          average_precision))


from sklearn.metrics import precision_recall_curve
credit_curve=precision_recall_curve(y_test,pred)
print(credit_curve)
```
